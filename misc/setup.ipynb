```
import re
```

```
import pandas as pd
list_data = df.select("_corrupt_record").collect()
print("Done Collection")
cols = ['activationTimestamp','expiryTimestamp','productName','origin":','originName','destination":','destinationName','purchaseTimestamp','activationTimestamp']
def fun(data):
  l = []
  for record in data:
    record = str(record)
    values = record.split(',')
    d = {}
    for v in values: 
      for c in cols:
        if c in v:
          try:
            val = v.split(':')[1]
            val = re.sub('[^0-9a-zA-Z]+', '', val)
            c = re.sub('[^0-9a-zA-Z]+', '', c)
            d[c] = val
          
          except:
            pass
      
      for c in cols:
        c = re.sub('[^0-9a-zA-Z]+', '', c)
        if c not in d.keys():
            d[c] = ''
    l.append(d)
  return l 

output = fun(list_data)
data_spark = spark.createDataFrame(Row(**x) for x in output)
data_spark.write.format("com.databricks.spark.csv").save("./data_output.csv",header = 'true')
```

Other option with Pandas Koalas -- this I think similar to the pandas we know

```
from databricks import koalas as kd

kTicketSales = kd.read_csv("wasbs://mtamnrhackathon@mtamnrdatastore.blob.core.windows.net/masabi-etix/masbi_etix_mnrticketsales_*_20200709.csv")
```

```
display(kTicketSales)
```

```
kSummaryTickets = kTicketSales.groupby('TransactionDate')['TransactionDate'].count().reset_index(name='count')
print(kSummaryTickets)
```

```
import seaborn as sns
df= kSummaryTickets.toPandas() #since seaborn doesn't know how to work with koalas pandas. This will pull all data to a single note. Might need to scale up our memory if our data is really big
sns.lineplot(x="TransactionDate", y="count", data=df)
```
