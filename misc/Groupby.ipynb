```python
import pandas as pd 
 
data = pd.read_csv("scanned_data_all.csv") 
```

```python
data.head()
```

```python
count_by_station = data.groupby("Origin Name")["Ticket Number"].count()
```

```python
count_by_station
```

```python
from datetime import datetime

data['Scan Timestamp'] = pd.to_datetime(data['Scan Timestamp'])
data.head()
```

```python
result_group = data.groupby([pd.Grouper(key='Scan Timestamp', freq='30min'), 'Origin Name'])['Ticket Number'].apply(list).reset_index()
```

```python
result_group
```

```python
result_count = data.groupby([pd.Grouper(key='Scan Timestamp', freq='30min'), 'Origin Name'])["Ticket Number"].count()
```

```python
result_count
```
